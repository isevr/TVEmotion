```python
import pandas as pd
import re
```

```python
df = pd.read_excel('dataset.xlsx')

def fix_df(text):
    text = re.sub(r'[^a-zA-Z\s]','',text)
    text = re.sub(r'\d+', '', text)
    return text.lower()

df['utterance'] = df.utterance.apply(fix_df)
df['emotion'] = df.emotion.apply(lambda x: x.lower())
```

```python
df.to_csv('emodata.csv',index=False)
```
